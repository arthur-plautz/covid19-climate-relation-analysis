```python
df_limpo_idade_sc = df_limpo_sc[df_limpo_sc.idade.astype(float) < 108]
```

# ---------------------------------------------

```python
df_final = df_limpo_idade_sc[(df_limpo_idade_sc['evolucaoCaso'] != 'Cancelado') & (df_limpo_idade_sc['classificacaoFinal'] != 'Descartado')]
```

```python
df_final.columns
```

```python
df_final.tipoTeste.unique()
```

```python
df_final = df_final[(df_final.tipoTeste == 'RT-PCR') | (df_final.tipoTeste == 'TESTE RÁPIDO - ANTICORPO') | (df_final.tipoTeste == 'TESTE RÁPIDO - ANTÍGENO')]
```

```python
len(df_final)
```

```python
df_municipio = df_final.groupby('municipio').count().reset_index()
```

```python
df_municipio
```

```python
media_casos_municipio = df_municipio.id.sum() / len(df_municipio.municipio)
```

```python
df_municipio_media_acima = df_municipio[df_municipio.id > media_casos_municipio]
```
